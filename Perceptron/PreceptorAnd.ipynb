# Perceptor And XOR

## Importing the neccessary libraries

```python
import numpy as np
```

## Preparing the training Data

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

## Initializing the parameters

```python
# initializing weights and bias
w = np.zeros(2)
b = 0
# learning rate
learning_rate = 0.1
epochs = 10

print("Initial weights:", w)
print("Initial bias:", b)
```

## Defining the Activtion function (Step Function)

```python
# step fuction
def step_function(x):
    return 1 if x >= 0 else 0
```

## Training loop

```python
# Training loop
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute Weighted sum
        z = np.dot(w, X[i]) + b

        # 2. Apply Activation function
        y_pred = step_function(z)

        # 3. Compute Error
        error = y[i] - y_pred

        # 4. Update Weight
        w = w + learning_rate * error * X[i]

        # 5. Update Bias
        b = b + learning_rate * error

        print(f"Data Point: {X[i]}, True Label: {y[i]}, Predicted: {y_pred}, Error: {error}, Updated Weight: {w}, Updated Bias: {b}")
```

## Final Model Parameters

```python
# final weights and bias
print("\nFinal weights:", w)
print("Final bias:", b)
```

## Testing the Model

```python
# Testing the trained model
test_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
print("\nTesting the trained model:")
for i in range(len(X)):
    z = np.dot(w, test_data[i]) + b
    y_pred = step_function(z)
    print(f"Data Point: {X[i]}, True Label: {y[i]}, Predicted: {y_pred}")
```

```python
# plotting the decision boundary
import matplotlib.pyplot as plt
# Create a grid of points
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
# Compute the decision boundary
Z = np.array([step_function(np.dot(w, np.array([x, y])) + b) for x, y in zip(xx.ravel(), yy.ravel())])
Z = Z.reshape(xx.shape)
# Plotting
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.grid()
plt.title('Perceptron Decision Boundary for AND Logic Gate')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.show()
```

```python
## for XOR logic gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
# Training loop for XOR logic gate
w = np.zeros(2)
b = 0

print("\nTraining for XOR Logic Gate")
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step_function(z)
        error = y[i] - y_pred
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error
        print(f"Data Point: {X[i]}, True Label: {y[i]}, Predicted: {y_pred}, Error: {error}, Updated Weight: {w}, Updated Bias: {b}")
```

```python
# final weights and bias for XOR logic gate
print("\nFinal weights for XOR Logic Gate:", w)
print("Final bias for XOR Logic Gate:", b)
```

```python
# testing the trained model for XOR logic gate
print("\nTesting the trained model for XOR Logic Gate:")
for i in range(len(X)):
    z = np.dot(w, X[i]) + b
    y_pred = step_function(z)
    print(f"Data Point: {X[i]}, True Label: {y[i]}, Predicted: {y_pred}")
```

```python
# plotting the decision boundary for XOR logic gate
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = np.array([step_function(np.dot(w, np.array([x, y])) + b) for x, y in zip(xx.ravel(), yy.ravel())])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.grid()
plt.title('Perceptron Decision Boundary for XOR Logic Gate')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.show()
```
