## Importing the neccessary libraries

```python
import numpy as np
import pandas as pd
```

## Step 1 : Prepare Training Data

```python
# Training Data
X = np.array([1,2,3,4])   # Study Hours
y = np.array([0,0,1,1])   # Fail(0)/Pass(1)

X,y
```

## Step 2 : Initialize parameters 

```python
w = 0 # weight
b = 0  # bias
learning_rate = 1
epochs = 5

print("Initial weight:", w)
print("Initial bias:", b)
```

## Step 3 : Defining the Activation function (Step Function)

```python
def step_function(x):
    return 1 if x >= 0 else 0
```

## Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute Weighted sum
        z = w * X[i] + b

        # 2. Apply Activation function
        y_pred = step_function(z)

        # 3. Compute Error
        error = y[i] - y_pred

        # 4. Update Weight
        w = w + learning_rate * error * X[i]

        # 5. Update Bias
        b = b + learning_rate * error

        print(f"Data Point: {X[i]}, True Label: {y[i]}, Predicted: {y_pred}, Error: {error}, Updated Weight: {w}, Updated Bias: {b}")
```

## Final Model Parameters

```python
# final model parameters
print("\nFinal weight:", w)
print("Final bias:", b)
```

## Testing the Model

```python
# testing the model
test_data = np.array([
    [1.5],  # should predict 0 (fail)
    [3.5],  # should predict 1 (pass)
    [0.5],  # should predict 0 (fail)
    [4.5]   # should predict 1 (pass)
])
Z = w * test_data + b
predictions = np.array([step_function(z) for z in Z])
print("\nTest Predictions:")
for i in range(len(test_data)):
    print(f"Study Hours: {test_data[i][0]}, Predicted: {predictions[i]}")
```
