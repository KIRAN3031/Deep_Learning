# Diabetes Risk Prediction (Binary Classification)

    Neural Network Architecture:
    2 inputs -> 3 Hidden neurons (Sigmoid) -> 1 Ouptut(Sigmoid)
    Output -> 1 : high Risk
    Output -> 0 : Low Risk

## Importing the necessary libraries

```python
import numpy as np
```

```python
np.random.seed(0)
X = np.array([[120,85,150,95], # Glucose
               [25.6, 28.9, 30.1, 27.4] ## BMI
])
```

## Normalize Values

```python
X = X / 200
```

## Output : 1-> High Risk, 0 -> Low Risk

```python
y = np.array([[1, 0, 1, 0]]) # Diabetes (1) or No Diabetes (0)
```

```python
X.shape, y.shape
```

## Sigmoid function

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

## Initializing the Parameters
    Hidden layer (3 neurons, 2 inputs)

```python
# Initialize weights and biases for hidden layer
w1 = np.random.rand(3, 2)*0.1 # 3 neurons in hidden layer, 2 features in input layer
b1 = np.random.rand(3, 1) # 3 neurons in hidden layer
# Initialize weights and biases for output layer
w2 = np.random.rand(1, 3)*0.1 # 1 neuron in output layer, 3 neurons in hidden layer
b2 = np.random.rand(1, 1) # 1 neuron in output layer
```

## Forward Propagation

```python
# hidden layer forward pass
z1 = np.dot(w1, X) + b1
a1 = sigmoid(z1)

# output layer forward pass
z2 = np.dot(w2, a1) + b2
a2 = sigmoid(z2)

print("Predicted probabilities (a2):", a2)
```

## Loss Function

### Binary Cross Entropy

```python
m = y.shape[1] # number of examples
loss = -np.sum(y * np.log(a2) + (1 - y) * np.log(1 - a2)) / m
print("Loss:", loss)
```

## Back Propagation
    output layer Gradient

```python
dZ2 = a2 - y
dW2 = np.dot(dZ2, a1.T) / m
db2 = np.sum(dZ2, axis=1, keepdims=True) / m
```

## Hidden layer Gradient
    Sigmoid derviatvie:

```python
dA1 = np.dot(w2.T, dZ2)
dZ1 = dA1 * a1 * (1 - a1)

dW1 = np.dot(dZ1, X.T) / m
db1 = np.sum(dZ1, axis=1, keepdims=True) / m
```

## Update Parameters

```python
learning_rate = 0.5
w2 -= learning_rate * dW2
b2 -= learning_rate * db2

w1 -= learning_rate * dW1
b1 -= learning_rate * db1
```

## Complete Training Loop

```python
epochs = 5000
for i in range(epochs):
    # Forward pass
    z1 = np.dot(w1, X) + b1
    a1 = sigmoid(z1)
    
    z2 = np.dot(w2, a1) + b2
    a2 = sigmoid(z2)
    
    # Compute loss
    loss = -np.sum(y * np.log(a2) + (1 - y) * np.log(1 - a2)) / m
    
    # Backward pass
    dZ2 = a2 - y
    dW2 = np.dot(dZ2, a1.T) / m
    db2 = np.sum(dZ2, axis=1, keepdims=True) / m
    
    dA1 = np.dot(w2.T, dZ2)
    dZ1 = dA1 * a1 * (1 - a1)
    
    dW1 = np.dot(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m
    
    # Update weights and biases
    w2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    w1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    
    if i % 500 == 0:
        print(f"Epoch {i}, Loss: {loss:.4f}")
```
