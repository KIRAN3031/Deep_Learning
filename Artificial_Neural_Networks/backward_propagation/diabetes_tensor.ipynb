# Diabetes Risk Prediction (Binary Classification)

    Neural Network Architecture:
    2 inputs -> 3 Hidden neurons (Sigmoid) -> 1 Ouptut(Sigmoid)
    Output -> 1 : high Risk
    Output -> 0 : Low Risk

## Importing the neccessary libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
np.random.seed(0)
X = np.array([[120,85,150,95], # Glucose
               [25.6, 28.9, 30.1, 27.4] ## BMI
])
X = X / 200 # Normalize features

y = np.array([[1, 0, 1, 0]]) # 1: diabetic, 0: non-diabetic

X.shape, y.shape
```

## Build Neural Network Model


```python
model = Sequential([
    Dense(3, input_dim =2 ,activation='sigmoid'), # Hidden layer with 3 neurons 2 input features
    Dense(1, activation='sigmoid') # Output layer with 1 neuron
])
```

## Compile the Model

```python
# Compile the model
model.compile(optimizer='sgd', loss=BinaryCrossentropy(), metrics=['accuracy'])
```

## Train the Model

```python
# train the model
model.fit(X.T, y.T, epochs=1000, verbose=0)
```

## Make Predictions

```python
# Making predictions
predictions = model.predict(X.T)
print("Predicted probabilities:\n", predictions)
```
