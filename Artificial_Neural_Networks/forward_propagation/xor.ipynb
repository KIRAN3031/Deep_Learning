## XOR Problem using multi layer neural Network (Forward Progation)

### Importing the neccessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Defining XOR Dataset

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
X , y
```

### Defining the Step Function

```python
def step(z):
    return 1 if z >= 0 else 0
```

### Initialize Weigths and Biases (Manually Chosen)

```python
# Hidden layer weights and bias
w1, w2, b1 =1, 1, -0.5
w3, w4, b2 =1, 1, -1.5

# Output layer weights and bias
w5, w6, b3 =1, -2, -0.5

print("Hidden Layer:")
print(f"Neuron 1: w1={w1}, w2={w2}, b1={b1}")
print(f"Neuron 2: w3={w3}, w4={w4}, b2={b2}")
print("\nOutput Layer:")
print(f"Output Neuron: w5={w5}, w6={w6}, b3={b3}")
```

### Forward Propagation

```python
def forward_propagation(x1, x2):
    # Hidden layer computations
    z1 = w1 * x1 + w2 * x2 + b1
    z2 = w3 * x1 + w4 * x2 + b2
    a1 = step(z1)
    a2 = step(z2)
    
    # Output layer computation
    y = w5 * a1 + w6 * a2 + b3
    output = step(y)
    
    return output
```

## Testing the forward progation for ALL XOR inputs

```python
# test the network on all input combinations with actual values
for x1, x2 in X:
    output = forward_propagation(x1, x2)
    print(f"Input: ({x1}, {x2}) -> Output: {output}")
```

### Visualization

```python
# XOR logic gate graph distribution
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y.flatten(), palette='Set1')
plt.title('XOR Logic Gate Distribution')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.legend(title='Output')
plt.grid()
plt.show()
```
